```python
import os
from dlup import SlideImage
import matplotlib.pyplot as plt
import numpy as np
```

```python
SCRATCH = '/data/archief/AMC-data/Barrett/experiments/LANS_tissue_seg_ahcore/segmentation_inference/2024-02-20_12-35-24/outputs/UNet/step_0'
DATA_DIR = '/data/archief/AMC-data/Barrett/LANS'
```

```python
f = open(os.path.join(SCRATCH, 'image_tiff_link.txt'))

for line in f.readlines():
    image_path = line.split(',')[0]
    mask_path = line.split(',')[1][:-1]
    print(image_path)
    print(mask_path)
    
    slide_image = SlideImage.from_file_path(image_path, )
    mask_image = SlideImage.from_file_path(mask_path)
    
    image_pil = slide_image.get_thumbnail()
    mask_pil = mask_image.get_thumbnail()
    
    np_image = np.asarray(image_pil)
    np_mask = np.asarray(mask_pil)
    
    plt.figure(figsize=(10, 6))
    plt.imshow(np_image)
    plt.show()
    plt.figure(figsize=(10, 6))
    plt.imshow(np_mask * 255)
```
